```python
# Import the random module
import random

# Define the grid as a list of lists
grid = [[" ", " ", " "],
        [" ", " ", " "],
        [" ", " ", " "]]

# Define the Q-table as a dictionary of dictionaries
Q_table = {}

# Define the learning rate
alpha = 0.1

# Define the discount factor
gamma = 0.9

# Define the exploration rate
epsilon = 0.1

# Define a function to print the grid
def print_grid():
    print("-------------")
    for i in range(3):
        print("|", end=" ")
        for j in range(3):
            print(grid[i][j], end=" | ")
        print()
        print("-------------")

# Define a function to check if the grid is full
def is_full():
    for i in range(3):
        for j in range(3):
            if grid[i][j] == " ":
                return False
    return True

# Define a function to check if a player has won
def has_won(player):
    # Check horizontal rows
    for i in range(3):
        if grid[i][0] == grid[i][1] == grid[i][2] == player:
            return True
    # Check vertical columns
    for j in range(3):
        if grid[0][j] == grid[1][j] == grid[2][j] == player:
            return True
    # Check diagonal lines
    if grid[0][0] == grid[1][1] == grid[2][2] == player:
        return True
    if grid[0][2] == grid[1][1] == grid[2][0] == player:
        return True
    # No winner
    return False

# Define a function to get the human player's move
def get_human_move():
    while True:
        # Get the row and column from the user
        row = int(input("Enter the row (1-3): ")) - 1
        col = int(input("Enter the column (1-3): ")) - 1
        # Check if the input is valid
        if row in range(3) and col in range(3):
            # Check if the cell is empty
            if grid[row][col] == " ":
                # Mark the cell with O
                grid[row][col] = "O"
                # Break the loop
                break
            else:
                # The cell is already occupied
                print("The cell is not empty. Try again.")
        else:
            # The input is out of range
            print("Invalid input. Try again.")

# Define a function to get the learning agent's move
def get_agent_move():
    # Get the current state as a string
    state = "".join(["".join(row) for row in grid])
    # Check if the state is in the Q-table
    if state not in Q_table:
        # Initialize the Q-values for each action to zero
        Q_table[state] = {(i, j): 0 for i in range(3) for j in range(3) if grid[i][j] == " "}
    # Choose an action using an epsilon-greedy strategy
    if random.random() < epsilon:
        # Choose a random action
        action = random.choice(list(Q_table[state].keys()))
    else:
        # Choose the action with the highest Q-value
        action = max(Q_table[state], key=Q_table[state].get)
    # Return the action as a tuple
    return action
# Define a function to update the Q-table based on the action and the reward
def update_Q_table(action, reward):
    # Get the current state as a string
    state = "".join(["".join(row) for row in grid])
    # Get the current Q-value for the action
    Q_value = Q_table[state][action]
    # Get the next state as a string
    i, j = action
    grid[i][j] = "X"
    next_state = "".join(["".join(row) for row in grid])
    grid[i][j] = " "
    # Check if the next state is in the Q-table
    if next_state not in Q_table:
        # Initialize the Q-values for each action to zero
        Q_table[next_state] = {(i, j): 0 for i in range(3) for j in range(3) if grid[i][j] == " "}
    # Get the maximum Q-value for the next state
    max_Q_value = max(Q_table[next_state].values())
    # Apply the Q-learning formula to update the Q-value
    Q_value = Q_value + alpha * (reward + gamma * max_Q_value - Q_value)
    # Update the Q-table with the new Q-value
    Q_table[state][action] = Q_value
```
